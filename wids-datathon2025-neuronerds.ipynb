# üß† 8th Annual WiDS Datathon: Unraveling the Mysteries of the Female Brain

## Project Summary

This notebook presents an overview of our contribution to the **8th Annual Women in Data Science (WiDS) Datathon**, focused on analyzing brain activity patterns associated with **ADHD (Attention-Deficit/Hyperactivity Disorder)** and exploring **how these patterns differ between males and females**. The central dataset in this challenge comprises **functional MRI (fMRI) connectome matrices**, capturing dynamic brain activity, complemented by rich metadata including socio-demographic factors, emotional assessments, and parenting information.

---

## üîç Core Challenge

> **‚ÄúWhat brain activity patterns are associated with ADHD? Are they different between males and females, and, if so, how?‚Äù**

Our objective is two-fold:
1. **Predict ADHD diagnosis** using neuroimaging and behavioral data.
2. **Model sex-based differences** in brain activity patterns, particularly in individuals diagnosed with ADHD.

---

## üì¶ Data Overview

The dataset includes:
- Functional connectome matrices derived from fMRI scans
- Subject-level metadata: age, sex, emotions, behavioral and parenting scores
- Labels: ADHD diagnosis and biological sex

This high-dimensional and multimodal dataset presents both challenges and opportunities for building robust predictive models and uncovering underlying neurological patterns.
